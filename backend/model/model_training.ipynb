```python
def remove_outliers(df,name):
  Q3=df[name].quantile(0.75)
  Q1=df[name].quantile(0.25)
  IQR=Q3-Q1
  upper_limit=Q3+(1.5*IQR)
  lower_limit=Q1-(1.5*IQR)
  return df[(df[name] < upper_limit) & (df[name] > lower_limit)]

```
